```python
import tkinter as tk
from tkinter import messagebox
import random
import string

def generate_password():
    length = length_var.get()
    use_digits = digits_var.get()
    use_special = special_var.get()

    if length < 4:
        messagebox.showerror("Error", "Password length should be at least 4.")
        return

    characters = string.ascii_letters
    if use_digits:
        characters += string.digits
    if use_special:
        characters += string.punctuation

    if not characters:
        messagebox.showerror("Error", "No character types selected.")
        return

    password = [
        random.choice(string.ascii_lowercase),
        random.choice(string.ascii_uppercase),
    ]
    if use_digits:
        password.append(random.choice(string.digits))
    if use_special:
        password.append(random.choice(string.punctuation))

    while len(password) < length:
        password.append(random.choice(characters))

    random.shuffle(password)
    password_str = ''.join(password)
    password_entry.delete(0, tk.END)
    password_entry.insert(0, password_str)

def copy_password():
    password = password_entry.get()
    if password:
        root.clipboard_clear()
        root.clipboard_append(password)
        messagebox.showinfo("Copied", "Password copied to clipboard.")
    else:
        messagebox.showwarning("No Password", "Generate a password first.")

# Create window
root = tk.Tk()
root.title("Random Password Generator")
root.geometry("400x250")
root.resizable(False, False)

# Variables
length_var = tk.IntVar(value=12)
digits_var = tk.BooleanVar(value=True)
special_var = tk.BooleanVar(value=True)

# UI Layout
tk.Label(root, text="Random Password Generator", font=("Helvetica", 16)).pack(pady=10)

tk.Label(root, text="Password Length:").pack()
tk.Spinbox(root, from_=4, to=64, textvariable=length_var, width=5).pack()

tk.Checkbutton(root, text="Include Digits (0-9)", variable=digits_var).pack()
tk.Checkbutton(root, text="Include Special Characters (!@#...)", variable=special_var).pack()

password_entry = tk.Entry(root, font=("Courier", 14), justify="center")
password_entry.pack(pady=10, fill=tk.X, padx=20)

tk.Button(root, text="Generate Password", command=generate_password).pack(pady=5)
tk.Button(root, text="Copy to Clipboard", command=copy_password).pack()

# Run app
root.mainloop()
```
